```python
import sys, os
sys.path.append(os.path.abspath('..'))

import pandas as pd
import mlflow
from src.mlflow_setup import init_tracking

init_tracking()

model_uri = "models:/walmart_sales_best_model/2"
model = mlflow.pyfunc.load_model(model_uri)

print("Model loaded")
```

```python
test = pd.read_csv('../data/test.csv')
print(test.shape)
test.head()
```

```python
predictions = model.predict(test)
print(predictions[:10])
```

```python
submission = pd.DataFrame({
    'Id': test['Store'].astype(str) + '_' + test['Dept'].astype(str) + '_' + test['Date'].astype(str),
    'Weekly_Sales': predictions
})

submission.head()
```

```python
submission.to_csv('../data/submission.csv', index=False)
print("Saved submission.csv")
```

```python
mlflow.set_experiment("XGBoost_Training")

with mlflow.start_run(run_name="XGBoost_kaggle_submission"):
    mlflow.log_param("model_version", 1)
    mlflow.log_param("submission_file", "submission.csv")
    mlflow.log_artifact('../data/submission.csv')
    print("Submission logged")
```
